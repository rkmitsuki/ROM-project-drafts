# 0. Import Data

```python
import pandas as pd
import numpy as np
from sklearn.model_selection import train_test_split
import tensorflow as tf
tf.config.run_functions_eagerly(True)
```

```python
df = pd.read_csv('datacfd.csv')
```

```python
df = df[["Strut wedge angle","Combustion chamber height","Top wall distance","Mach at outlet chamber","Shock angle","Slope after shock", "Slope chamber","Total Pressure","Total Temperature","Performace efficiency","Slope between shock and chamber"]]
X = pd.get_dummies(df.drop(["Strut wedge angle","Combustion chamber height","Top wall distance"], axis=1))
y = pd.get_dummies(df.drop(["Mach at outlet chamber","Shock angle","Slope after shock","Slope chamber","Total Pressure","Total Temperature","Performace efficiency","Slope between shock and chamber"], axis=1))
```

```python
X_train, X_test, y_train, y_test = train_test_split(X, y, test_size=.2)
```

# 1. Import Dependencies

```python
from tensorflow.keras.models import Sequential, load_model
from tensorflow.keras.layers import Dense
from sklearn.metrics import accuracy_score
```

# 2. Build and Compile Model

```python
model = Sequential()
model.add(Dense(units=64, activation='relu', input_dim=len(y_train.columns)))
model.add(Dense(units=64, activation='relu'))
model.add(Dense(units=64, activation='relu'))
model.add(Dense(units=8))
```

```python
model.compile(optimizer='adam', loss='mean_squared_error', metrics=['mean_squared_error'])
```

# 3. Fit, Predict and Evaluate

```python
X_test = np.asarray(X_test).astype(np.float16)
y_test = np.asarray(y_test).astype(np.float16)
X_train = np.asarray(X_train).astype(np.float16)
y_train = np.asarray(y_train).astype(np.float16)
model.fit(y_train, X_train, epochs=1500, batch_size=64)
```

```python
y_hat = model.predict(y_test)
```
